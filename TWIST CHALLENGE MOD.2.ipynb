Columns not needed will be dropped first before we proceed with data cleaning

```python
%%sql

ALTER TABLE salaries

DROP COLUMN salary_id,
DROP COLUMN employer_name, 
DROP COLUMN location_name,
DROP COLUMN location_country,
DROP COLUMN location_state,
DROP COLUMN location_latitude, 
DROP COLUMN location_longitude,
DROP COLUMN job_title_category,
DROP COLUMN job_title_rank, 
DROP COLUMN employer_experience_years,
DROP COLUMN stock_value_bonus, 
DROP COLUMN comments, 
DROP COLUMN submitted_at
```

Check for nulls in the columns

```python
%%sql

SELECT *

FROM salaries 

WHERE total_experience_years IS NULL
```

since its a large data base we will delete the nulls

```python
%%sql

DELETE FROM salaries

WHERE total_experience_years IS NULL
```

Lets verify the nulls have been deleted if it prints out 0, it means they have been deleted

```python
%%sql

SELECT *

FROM salaries 

WHERE total_experience_years IS NULL
```

Since the annual base pay is the requirement for every salaried worker, we check both for NULLS and 0, Nulls in SQL are outputed as none and hence it would be impossible to perform arithmetic calculations where NULLS are. If any is found, the rows will be deleted.

```python
%%sql

SELECT *

FROM salaries

WHERE annual_base_pay IS NULL
```

```python
%%sql

SELECT *

FROM salaries

WHERE annual_base_pay = 0
```

For signing bonus and annual bonus, we only check for the NULLS because not every employee receives a signing bonus or annual bonus.

```python
%%sql

SELECT *

FROM salaries

WHERE signing_bonus IS NULL
```

```python
%%sql

SELECT *

FROM salaries

WHERE annual_bonus IS NULL
```

We now add a new column **total compensation** since we are investigating the relationship between years of experience and salary

```python
%%sql

ALTER TABLE salaries

ADD COLUMN total_compensation integer
```

View to see column

```python
%%sql

SELECT *

FROM salaries

LIMIT 5
```

Next, we calculate the values for the total compensation column.

```python
%%sql
    
UPDATE salaries

SET total_compensation=annual_base_pay+signing_bonus+annual_bonus
```

View to see the calculation has been performed

```python
%%sql

SELECT *

FROM salaries

LIMIT 5
```

Now that we have our filtered table, we will store our query in a pandas DataFrame, and then build our regression model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
import seaborn as sns
warnings.filterwarnings("ignore")
```

```python
salaries_df = pd.read_sql('SELECT * FROM salaries', engine)
salaries_df.head()
```

```python
salaries_df.dtypes
```

```python
salaries_df['total_experience_years'] = salaries_df['total_experience_years'].astype(float)
```

```python
salaries_df.dtypes
```

Dropping the 'annual_base_pay','signing_bonus','annual_bonus' columns as only the aggregated column 'total_compensation' will be used going forward.

```python
salaries_df.drop(['annual_base_pay','signing_bonus','annual_bonus'],axis=1,inplace=True)
```

```python
salaries_df.head()
```

```python
#salaries_df['total_compensation']=salaries_df['total_compensation'].div(10000)
```

```python
salaries_df.head()
```

We check the correlation between total years experience and total compensation. Even though there is a positive relationship between the two, the relationship is not a strong one per our data. 

```python
salaries_df.corr()
```

```python
sns.heatmap(salaries_df.corr(),annot=True,lw=1)
```

```python
x = salaries_df.iloc[:, 1].values.reshape(-1, 1)
y = salaries_df.iloc[:, -1].values.reshape(-1, 1)
```

```python
print(x)
```

```python
salaries_df['total_experience_years'].min()
```

```python
print(y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
# print the intercept
print(model.intercept_)
```

```python
import pickle

filename = 'finalized_model.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
import pandas as pd
import plotly.express as px 

import dash 
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output

app = dash.Dash(__name__)

df = pd.read_sql('SELECT * FROM salaries', engine)

app.layout = html.Div([
    html.H4('Salary v Years of Experience', style={'text-align': 'center'}),
    dcc.Slider(1,40,1,
               value=10,
               id='yrs_of_exp'
),
    html.Div(id='output_container', children=[]),
    html.Br(),

    dcc.Graph(id='predict_graph', figure={})

])


@app.callback(
    [Output(component_id='output_container', component_property='children'),
    Output(component_id='predict_graph', component_property='figure')],
    [Input(component_id='yrs_of_exp', component_property='value')]
)


def update_graph(option_slctd):
    print(option_slctd)
    print(type(option_slctd))

    container = "The years of experience chosen by user was: {}".format(option_slctd)

    dff = df.copy()
    dff = dff[dff["total_experience_years"] == option_slctd]
    
    fig = px.scatter(df, x="total_experience_years", y="total_compensation")

        
    return container, fig
if __name__ == '__main__':
    app.run_server(debug=False)
```
